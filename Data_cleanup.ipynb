```python
data_201912 = pd.read_csv('Resources/201912-citibike-tripdata.csv')
data_201911 = pd.read_csv('Resources/201911-citibike-tripdata.csv')
data_201910 = pd.read_csv('Resources/201910-citibike-tripdata.csv')
data_201909 = pd.read_csv('Resources/201909-citibike-tripdata.csv')
data_201908 = pd.read_csv('Resources/201908-citibike-tripdata.csv')
data_201907 = pd.read_csv('Resources/201907-citibike-tripdata.csv')
data_201906 = pd.read_csv('Resources/201906-citibike-tripdata.csv')
data_201905 = pd.read_csv('Resources/201905-citibike-tripdata.csv')
data_201904 = pd.read_csv('Resources/201904-citibike-tripdata.csv')
data_201903 = pd.read_csv('Resources/201903-citibike-tripdata.csv')
data_201902 = pd.read_csv('Resources/201902-citibike-tripdata.csv')
data_201901 = pd.read_csv('Resources/201901-citibike-tripdata.csv')
```

```python
data_2019 = pd.concat([data_201901, data_201902, data_201903, data_201904, data_201905,\
                       data_201906, data_201907, data_201908, data_201909, data_201910,\
                       data_201911, data_201912], sort=False)
```

```python
len(data_2019)
```

```python
# Check number of rows per usertype and gender to help decide which group to analyze
data_2019.groupby(['usertype','gender']).count()
```

```python
# Check number of rows per "early birth year" to help decide if I eliminate most likely wrong data
data_2019.groupby(['birth year']).count()
```

```python
# Final decision:
#  - Analyze Customers with identified gender
#  - Analyze only customer with birth year > 1920 (assuming the rest are misspelled birth year)
```

```python
data_customer_2019 = data_2019[(data_2019['usertype'] == 'Customer') & (data_2019['gender'] != 0) & (data_2019['birth year'] > 1920)]
```

```python
len(data_customer_2019)
```

```python
data_customer_2019.to_csv('citibike_customers_2019.csv')
```
